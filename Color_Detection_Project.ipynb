```python
import cv2
import pandas as pd
import urllib.request
import os

img_path = 'image.jpg.jpg'
csv_path = 'colors.csv'

if not os.path.exists(csv_path):
    print("[INFO] Downloading colors dataset...")
    url = 'https://raw.githubusercontent.com/amankharwal/Website-data/master/colors.csv'
    try:
        urllib.request.urlretrieve(url, csv_path)
        print("[INFO] Download complete!")
        
    except Exception as e:
        print("error uploading data check your internet")
        exit()

img = cv2.imread(img_path)
if img is None:
    print("check that you put a photo name image.jpg.jpg can't find it")
    exit()

img = cv2.resize(img, (800, 600))

index = ["color", "color_name", "hex", "R", "G", "B"]
csv = pd.read_csv(csv_path, names=index, header=None)

clicked = False
r = g = b = x_pos = y_pos = 0

def get_color_name(R, G, B):
    minimum = 10000
    cname = ""
    for i in range(len(csv)):
        d = abs(R - int(csv.loc[i, "R"])) + abs(G - int(csv.loc[i, "G"])) + abs(B - int(csv.loc[i, "B"]))
        if d <= minimum:
            minimum = d
            cname = csv.loc[i, "color_name"]
    return cname

def draw_function(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDBLCLK: 
        global b, g, r, x_pos, y_pos, clicked
        clicked = True
        x_pos = x
        y_pos = y
        b, g, r = img[y, x]
        b = int(b)
        g = int(g)
        r = int(r)

cv2.namedWindow('Color Detection')
cv2.setMouseCallback('Color Detection', draw_function)

print("[INFO] System Ready! Double-click anywhere on the image.")

while True:
    cv2.imshow("Color Detection", img)
    if clicked:
        cv2.rectangle(img, (20, 20), (750, 60), (b, g, r), -1)
        text = get_color_name(r, g, b) + ' R=' + str(r) + ' G=' + str(g) + ' B=' + str(b)

        text_color = (255, 255, 255)
        if r + g + b >= 600:
            text_color = (0, 0, 0)

        cv2.putText(img, text, (50, 50), 2, 0.8, text_color, 2, cv2.LINE_AA)
        clicked = False

    if cv2.waitKey(20) & 0xFF == ord('q'):
        break

cv2.destroyAllWindows()
```
